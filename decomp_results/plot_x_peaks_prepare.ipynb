```python
def rotate_via_numpy(x, y, radians):
    """Use numpy to build a rotation matrix and take the dot product."""
    #print (x)
    c, s = np.cos(radians), np.sin(radians)
    j = np.matrix([[c, s], [-s, c]])
    m = np.dot(j, [x, y])
    #print (m.T[:,0])
    #print (m)
    return np.array(m.T[:,0]), np.array(m.T[:,1])
```
