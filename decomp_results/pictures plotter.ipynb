```python
from scipy.ndimage import rotate

def slice_rotate(img, angle, y, x1, x2):
    resid = rotate(img, angle=angle, reshape=False)

    fig = plt.figure(figsize=[12,12])
    plt.subplot(2, 1, 1)
    plt.imshow(resid, norm=SymLogNorm(linthresh=0.03, vmin=-0.3, vmax=0.3), cmap=cm.RdBu)
    plt.plot([x1, x2], [y, y], color='k')
    plt.title('Residual image')

    plt.subplot(2, 1, 2)
    x_slice = resid[y, :]
    plt.xlim(x1, x2)
    plt.plot(x_slice, '.', color='k', label='Image')
    plt.yscale('log')
    plt.show()
```
